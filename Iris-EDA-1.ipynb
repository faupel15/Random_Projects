## Load the Dataset

```python
# Load the Iris dataset from seasborn library
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
iris = sns.load_dataset('iris')
```

# Data Cleaning

```python
# Check for missing values
iris.isnull().sum()
```

# Data Visualization

```python
# Plot Sepal Length
ax = sns.histplot(data=iris, x='sepal_length', kde=True)
ax.set_title('Distribution of Sepal Length')
ax.set_xlabel('Sepal Length (cm)')
plt.show()
```

```python
# Plot Sepal Width
ax = sns.histplot(data=iris, x='sepal_width', kde=True)
ax.set_title('Distribution of Sepal Width')
ax.set_xlabel('Sepal Width (cm)')
plt.show()
```

```python
# Plot Petal Length
ax = sns.histplot(data=iris, x='petal_length', kde=True)
ax.set_title('Distribution of Petal Length')
ax.set_xlabel('Petal Length (cm)')
plt.show()
```

```python
# Plot Petal Width
ax = sns.histplot(data=iris, x='petal_width', kde=True)
ax.set_title('Distribution of Petal Width')
ax.set_xlabel('Petal Width (cm)')
plt.show()
```

```python
# Pair plots of numeric variables
ax = sns.pairplot(iris, hue='species')
plt.suptitle('Relationships Among Variables in the Iris Dataset', y=1.05)  # y parameter is used to adjust the title's position.
plt.show()
```

```python
# Calculate and display the correlation matrix
correlation_matrix = iris.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Numeric Variables in the Iris Dataset')
plt.show()
```

```python
# Show the count of the different species using a barplot
ax = sns.countplot(data=iris, x='species')
ax.set_title('Number of Records per Species')
plt.show()
```

```python
# Another way of showing this using the barplot function 

# Prepare data
species_count = iris['species'].value_counts().reset_index()
species_count.columns = ['species', 'count']

# Plot
sns.barplot(data=species_count, x='species', y='count')
plt.title('Count of each Species in the Iris Dataset')
plt.show()
```

# Data Analysis

```python
# This was already done above, but here it is again for context

# Calculate and display the correlation matrix
correlation_matrix = iris.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Numeric Variables in the Iris Dataset')
plt.show()
```

- Petal length and petal width have the strongest positive correlation of all the features at 0.96, meaning that as petal length increases, petal width tends to increase proportionally. 
- Sepal length and petal length have a strong positive correlation of 0.87, meaning that as petal length increases, so does sepal length. 
- Sepal length also has a strong positive correlation with petal_width, at 0.82.

# Summary

- The iris dataset contains 5 columns and 150 rows.
- Four of the columns are numeric: petal_length, petal_width, sepal_length, and sepal_width.
- One of the columns is categorical: species.
- There are no missing values in the dataset, and each of the three species in the dataset has 50 observations.
- The columns with the strongest positive correlation are petal length and petal width, with r=0.96.
- The columns with the strongest negative correlation are sepal width and petal length, with r=-0.43.
- Of the three species, Setosa has the smallest petals.
- Versicolor have larger petals than Setosa, but not as big as Virginica.
